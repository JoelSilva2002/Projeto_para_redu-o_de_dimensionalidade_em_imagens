```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
#carrega imagem
img = Image.open("be337d8282085c92544c25b554c38836.webp")
plt.imshow(img)
plt.title("Imagem Original")
plt.axis('off')
plt.show()
```

```python
#Converter para tons de cinza
img_gray = img.convert("L")  # "L" = grayscale
plt.imshow(img_gray, cmap="gray")
plt.title("Imagem em Cinza")
plt.axis('off')
plt.show()
```

```python
#Converter para imagem binária
#Definindo threshold (ex: 128)
threshold = 128
img_binary = img_gray.point(lambda x: 255 if x > threshold else 0)
plt.imshow(img_binary, cmap="gray")
plt.title("Imagem Binária")
plt.axis('off')
plt.show()
```

```python
#transformar em array NumPy
arr_gray = np.array(img_gray)
arr_binary = np.array(img_binary)
print("Array Cinza:", arr_gray.shape)
print("Array Binário:", arr_binary.shape)
```
